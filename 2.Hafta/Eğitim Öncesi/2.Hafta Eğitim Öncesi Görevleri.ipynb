````python
# 3.Sorunun çözümünü buraya yazabilirsin.

# pi sayısını 3.14 olarak tanımlıyoruz
pi = 3.14

# Kullanıcıdan yarıçapı girmesini istiyoruz
yaricap = float(input("Lütfen dairenin yarıçapını giriniz: "))

# Dairenin alanını hesaplıyoruz
alan = pi * yaricap * yaricap

# Sonucu ekrana bastırıyoruz
print("Dairenin alanı: ", alan)
````

### Soru 4
x ve y adında iki liste oluşturunuz. Bu listeleri birbiryle yer değiştirerek yazdırınız.

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
``` 

```python	
Output:
x = [5, 4, 3, 2, 1]
y = [1, 2, 3, 4, 5]
```

````python
# İlk listeleri tanımlıyoruz
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Listeleri yer değiştiriyoruz
x, y = y, x

# Sonuçları ekrana bastırıyoruz
print("x =", x)
print("y =", y)
````
